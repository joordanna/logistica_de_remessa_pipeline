```python
import pyodbc

dados_conexao = (
    "Driver={SQL Server};"
    "Server=DESKTOP-0UN1LB3\\SQLEXPRESS;"
    "Database=LogisticsDB2;"
)

conexao = pyodbc.connect(dados_conexao)
cursor = conexao.cursor()
print('Conexão bem sucedida')
```

```python
#pegando os dataframes tratados. Salvo no notebook de tratamento
import pandas as pd

df_refinado = pd.read_csv("df_refinado.csv")
```

```python
colunas = df_refinado.columns.tolist()
column_names = ", ".join([f"[{col}]" for col in colunas])
placeholders = ", ".join(["?"] * len(colunas))

insert_query = f"""
INSERT INTO LogisticsTable ({column_names})
VALUES ({placeholders})
"""

values = df_refinado.itertuples(index=False, name=None)

cursor.executemany(insert_query, values)
conexao.commit()

print("Dados inseridos com sucesso!")
```
