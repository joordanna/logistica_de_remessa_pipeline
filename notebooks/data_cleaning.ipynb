```python
import pandas as pd

database = pd.read_csv('Dataset_Generator_for_DTDC.csv')
database
#database[:6]
```

```python
database.columns
```

```python
database.info()
```

```python
colunas_relevantes = [
    'Origin', 'Destination', 'Pouch No', 'Date', 'Consignment No',
    'Receive Date', 'Expiry Date',
    'Total Pieces', 'Actual Wt', 'Volumetric Wt', 'Chargeable Wt',
    'Tariff', 'VAS Charges', 'Total Amount',
    'Mode', 'Mode of Payment', 'Nature of Consignment', 'Risk Surcharge',
    'Sender City', 'Sender State', 'Sender Pincode',
    'Recipient City', 'Receiver State', 'Receiver Pincode'
]

df_refinado = database[colunas_relevantes].copy()
df_refinado.info()
```

```python
df_refinado[:6]
```

```python
df_refinado.isna().sum()
```

```python
colunas_datas = ['Date', 'Receive Date', 'Expiry Date']

df_refinado[colunas_datas] = df_refinado[colunas_datas].apply(pd.to_datetime, errors='coerce')
```

```python
df_refinado.info()
```

```python
df_refinado.dtypes
```

```python
df_refinado.columns = (
    df_refinado.columns
    .str.strip()              # tira espaços no começo/fim
    .str.replace(" ", "_")    # troca espaço por underline
)
```

```python
# Converter "nan", "None", "Null" e vazios para NaN real
df_refinado = df_refinado.replace(["nan", "None", "null", "Null", ""], pd.NA)

# Garantir que colunas float sejam float de verdade
colunas_float = [
    "Actual_Wt", "Volumetric_Wt", "Chargeable_Wt",
    "Tariff", "VAS_Charges", "Total_Amount"
]

for col in colunas_float:
    df_refinado[col] = pd.to_numeric(df_refinado[col], errors="coerce")
```

```python
df_refinado.to_csv("df_refinado.csv", index=False)
```
